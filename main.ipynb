```
import pandas as pd
import io
# récupération du fichier à traiter
# ici nous avons la colonne description qui présente partout et la colonne line_code qui contient des lignes "other"
# nous allons donc nous baser de la colonne description pour renseigner la colonne line_code
df = pd.read_csv(io.StringIO(uploaded['business-operations-survey-2018-business-finance-csv.csv'].decode('utf-8')))
df
```

```
# on va récupérer tous les descriptions dans le tableau models
# et tous les line_code dans le tableau constructeur
constructeur = []
models = []
for index, row in df.iterrows():
  if row.description in models:
    indexKey = models.index(row.description)
    _constructeur = constructeur[indexKey]
    if row.line_code not in _constructeur:
      _constructeur.append(row.line_code)
  else:
    models.append(row.description)
    constructeur.append([row.line_code])

    
```

```
# a present on itere sur notre df pour identifier les ligne dont le line_code est other et se baser sur les constructeurs du models
# correspondant pour setter le champs line_code
for index, row, in df.iterrows():
  if "other" in row.line_code:
    indexModel = models.index(row.description)
    current_contructeur = constructeur[indexModel]
    for construct in current_contructeur:
      if "other" not in construct:
        df.loc[index, 'line_code'] = construct

df
```
